```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Funzione per analisi rapida
def quick_analysis(df, title):
    print(f"--- {title} ---")
    print(df.describe(include='all'))
    print("\nMissing values per column:")
    print(df.isnull().sum())
    print("\n")

    # Grafico delle distribuzioni
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    for col in numeric_cols:
        plt.figure(figsize=(10, 4))
        sns.histplot(df[col].dropna(), kde=True)
        plt.title(f'{title}: Distribuzione di {col}')
        plt.show()
    
    # Pairplot per le prime 4 colonne numeriche (se ce ne sono almeno 4)
    if len(numeric_cols) >= 4:
        sns.pairplot(df[numeric_cols[:4]])
        plt.suptitle(f'{title}: Pairplot delle prime 4 colonne numeriche')
        plt.show()

# Importa i dataset sporchi
df_outliers = pd.read_csv(outliers_file_path)
df_inconsistents = pd.read_csv(inconsistents_file_path)
df_100_nulls = pd.read_csv(nulls_file_path)
df_dropped_column = pd.read_csv(dropped_column_file_path)

# Analizza i dataset
quick_analysis(df_outliers, 'Dataset con Outliers')
quick_analysis(df_inconsistents, 'Dataset con Valori Inconsistenti')
quick_analysis(df_100_nulls, 'Dataset con Colonna di Valori Nulli')
quick_analysis(df_dropped_column, 'Dataset con Colonna Rimossa')
```
